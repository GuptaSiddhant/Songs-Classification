```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.svm import SVC
from sklearn import svm
from sklearn.grid_search import GridSearchCV
```

```python
dfx = pd.read_csv("train_data.csv", header = None)
dfy = pd.read_csv("train_labels.csv" , header= None)
dftest =  pd.read_csv("test_data.csv", header = None)
L=[]
for m in dfy.as_matrix().tolist():
    L += m
    
X_train = dfx 
y_train = L
```

```python
np_scaled_test = preprocessing.quantile_transform(dftest)
#np_scaled_test = quantile_transform.fit_transform(dftest)
df_normalized_test = pd.DataFrame(np_scaled_test)
#df_normalized_test
df_normalized_test
```

```python
#min_max_scaler = preprocessing.MinMaxScaler()
#np_scaled = min_max_scaler.fit_transform(X_train)
#df_normalized = pd.DataFrame(np_scaled)
#df_normalized
```

```python
np_scaled = preprocessing.quantile_transform(X_train)
#np_scaled_test = quantile_transform.fit_transform(dftest)
df_normalized = pd.DataFrame(np_scaled)
#df_normalized_test
df_normalized
```

```python
parameter_candidates = [
  {'C': [1, 10, 100, 1000,10000], 'kernel': ['linear']},
  {'C': [1, 10, 100, 1000,10000], 'gamma': [0.1,0.01,0.001, 0.0001], 'kernel': ['rbf']},
]
```

```python
clf = GridSearchCV(estimator=svm.SVC(), param_grid=parameter_candidates, n_jobs=-1)
clf.fit(df_normalized, y_train) 
```

```python
print('Best score for data1:', clf.best_score_) 
print('Best C:',clf.best_estimator_.C) 
print('Best Kernel:',clf.best_estimator_.kernel)
print('Best Gamma:',clf.best_estimator_.gamma)
```

```python
#63.799
svm_model_linear_normalized = SVC(kernel = 'Linear', C = 1,gamma=0.001, probability= True).fit(df_normalized, y_train) 
```

```python
#65.919
svm_model_linear_normalized_rbf = SVC(kernel = 'rbf', C = 10,gamma=0.01, probability= True).fit(df_normalized, y_train) 
```

```python
svm_predictions_normalized_rbf = svm_model_linear_normalized_rbf.predict(df_normalized_test)
```
