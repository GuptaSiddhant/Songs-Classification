```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.svm import SVC
```

```python
dfx = pd.read_csv("train_data.csv", header = None)
dfy = pd.read_csv("train_labels.csv" , header= None)
dftest =  pd.read_csv("test_data.csv", header = None)
L=[]
for m in dfy.as_matrix().tolist():
    L += m
    
X_train = dfx 
y_train = L
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
np_scaled_test = min_max_scaler.fit_transform(dftest)
df_normalized_test = pd.DataFrame(np_scaled_test)
df_normalized_test
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
np_scaled = min_max_scaler.fit_transform(X_train)
df_normalized = pd.DataFrame(np_scaled)
df_normalized
```

```python
svm_model_linear_normalized = SVC(kernel = 'linear', C = 1, probability = True).fit(df_normalized, y_train) 
```

```python
svm_predictions_normalized = svm_model_linear_normalized.predict(dftest)
```

```python
print(svm_predictions_normalized)
```

```python
np.savetxt("svm_predictions_linear_normalized.csv", svm_predictions_normalized, delimiter=",")
```
